```python


# Extract the numeric part from the string
data['Unnamed: 0'] = data['Unnamed: 0'].str.extract('(\d+)')

# Convert the column to integer datatype
data['Unnamed: 0'] = data['Unnamed: 0'].astype(int)

# Print the DataFrame
print(data)
```

```python
data.index
```

```python
data['Unnamed: 0'] = data['Unnamed: 0'].astype(float)
```

```python
from sklearn.preprocessing import scale
```

```python
X = pd.DataFrame(scale(data), index = data.index, columns = data.columns)
```

```python
from sklearn.decomposition import PCA
```

```python
pc = PCA(n_components=10)
```

```python
pc_array = pc.fit_transform(X)
```

```python
pc_data = pd.DataFrame(pc_array, columns = ["pc1","pc2","pc3","pc4","pc5","pc6","pc7","pc8","pc9","pc10"])
```
